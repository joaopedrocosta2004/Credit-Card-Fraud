## Credit Card Fraud - Data Exploration

Imports

```
import pandas as pd
import numpy as np
import seaborn as sns
```

Load Dataset

```
df = pd.read_csv('/content/creditcard.csv')
display(df.head())
```

Data Understanding

```
print('Informações gerais do dataset:')
display(df.info())
```

```
print('\nEstatísticas descritivas do dataset:')
display(df.describe())
```

```
print('\nContagem de valores nulos por coluna:')
display(df.isnull().sum())
```

```
print('\nDistribuição da coluna "Class":')
display(df['Class'].value_counts())
display(df['Class'].value_counts(normalize=True).mul(100).round(2).astype(str) + '%')
```

```
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='Class', data=df)
ax.set_title('Distribuição de Transações Normais (0) e Fraudulentas (1)')
ax.set_xlabel('Classe da Transação')
ax.set_ylabel('Número de Transações')
ax.set_xticks(ticks=[0, 1], labels=['Normal (0)', 'Fraude (1)'])
plt.show()
```

```
plt.figure(figsize=(10, 6))
sns.kdeplot(df[df['Class'] == 0]['Amount'], label='Normal (0)', fill=True, alpha=0.5)
sns.kdeplot(df[df['Class'] == 1]['Amount'], label='Fraude (1)', fill=True, alpha=0.5)
plt.title('Distribuição do Valor da Transação por Classe')
plt.xlabel('Valor da Transação (Amount)')
plt.ylabel('Densidade')
plt.xlim(0, 2000)
plt.legend()
plt.show()
```

```
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the Hour column exists
df['Hour'] = (df['Time'] / 3600) % 24

# Create the visualization using Seaborn displot for overlapping KDE
sns.set_theme(style="whitegrid")
g = sns.displot(
    data=df,
    x="Hour",
    hue="Class",
    kind="kde",
    fill=True,
    common_norm=False,
    palette="tab10",
    alpha=0.5,
    height=6,
    aspect=2
)

# Adjust axis and titles
g.set_axis_labels("Hour of the Day (0 to 24h)", "Density")
g.ax.set_title("Transaction Distribution by Hour of the Day (Normal vs Fraud)")
g.ax.set_xticks(range(0, 25))
plt.show()
```
